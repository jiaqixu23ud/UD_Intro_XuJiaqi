```python
from pytube import YouTube
from pytube import Search

video_names = ['Rockit – Herbie Hancock', 'Thriller – Michael Jackson', 'Girls Just Want to Have Fun – Cyndi Lauper', 'Every Breath You Take – The Police', 'Dont Come Around Here No More – Tom Petty and the Heartbreakers', 'California Girls – David Lee Roth', 'Just a Gigolo / I Aint Got Nobody – David Lee Roth', 'We Are the World – USA for Africa', 'Take On Me – a-ha', 'Cry – Godley & Creme', 'Addicted to Love – Robert Palmer']

def get_video_url(video_name):
    try:
        search_query = video_name + ' YouTube'
        search = Search(search_query)
        video = search.results[0]
        return video.watch_url
    except Exception as e:
        print(f"Could not find URL for '{video_name}': {e}")
        return None

video_urls = {}
for video_name in video_names:
    url = get_video_url(video_name)
    if url:
        video_urls[video_name] = url

for video_name, url in video_urls.items():
    print(f"Video Name: {video_name} - URL: {url}")
```

```python
from pytube import YouTube

video_urls = {
    'https://youtube.com/watch?v=GHhD4PD75zY',
    'https://youtube.com/watch?v=sOnqjkJTMaA',
    'https://youtube.com/watch?v=PIb6AZdTr-A',
    'https://youtube.com/watch?v=OMOGaugKpzs',
    'https://youtube.com/watch?v=h0JvF9vpqx8',
    'https://youtube.com/watch?v=-JpM7i0yaz0',
    'https://youtube.com/watch?v=Tg2rrMwfgB0',
    'https://youtube.com/watch?v=9AjkUyX0rVw',
    'https://youtube.com/watch?v=djV11Xbc914',
    'https://youtube.com/watch?v=6a_PRcVYX7g',
    'https://youtube.com/watch?v=XcATvu5f9vE',
}

download_dir = 'D:\DATABASE\Music_Videos'

for url in video_urls:
    try:
        yt = YouTube(url)

        video_stream = yt.streams.get_highest_resolution()

        print(f"Downloading: {yt.title}")
        video_stream.download(download_dir)

        print(f"Download of '{yt.title}' complete.")
    except Exception as e:
        print(f"Could not download '{url}': {e}")

print("Batch download complete!")
```
