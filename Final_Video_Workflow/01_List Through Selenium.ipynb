# This is to abstract all the nominees of the MTV awards.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()

url = "https://en.wikipedia.org/wiki/MTV_Video_Music_Award_for_Video_of_the_Year"
driver.get(url)

wait = WebDriverWait(driver, 10)
element = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, '.wikitable.sortable')))

years = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr')
for year in years:
    cells = year.find_elements(By.TAG_NAME, 'td')
    if len(cells) > 2:
        nominees = cells[2].find_elements(By.XPATH, './/li')
        for index, nominee in enumerate(nominees):
            nominee_text = nominee.text
            if " – " in nominee_text:
                # Remove double quotes and replace newline with period and space
                formatted_nominee = nominee_text.replace('"', '').replace('\n', ', ').replace('†', '').replace('*', '')
                print(formatted_nominee)

driver.quit()
```
